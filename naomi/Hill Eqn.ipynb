```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams.update({'font.size': 20})  

def hill_equation(L, Kd, n):
    return (L**n) / (Kd + L**n)

L = np.logspace(-2, 2, 500)  
Kd = 1.0                    
n_values = [1, 2, 6, 26, 226]  

plt.figure(figsize=(8, 6))
for n in n_values:
    Y = hill_equation(L, Kd, n)
    plt.plot(L, Y, label=f"N={n}")
plt.xscale('log')
plt.xlabel('Protein Concentration')
plt.ylabel('Fraction Activated')
plt.legend()
plt.grid(True)
plt.show()
```
